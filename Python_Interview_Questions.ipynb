##### Cleaning of dataframe

```python
## Cleaning data using pandas.
import pandas as pd

df = pd.read_csv("employees.csv")

# Detect
df.isnull().sum()                          # count of nulls per column
df.isnull().mean() * 100                   # % missing per column

# Drop
df.dropna(subset=["salary", "department"]) # drop rows where key cols are null

# Fill — simple strategies
df["salary"].fillna(df["salary"].mean())           # mean imputation
df["department"].fillna(df["department"].mode()[0]) # mode for categoricals
df["salary"].fillna(method="ffill")                # forward fill (time series)
```

```python
# Scikit-learn — Imputation for ML Pipelines
'''When your goal is model training, dropping rows is wasteful and filling with mean/mode is often too naive. 
Scikit-learn provides statistically smarter imputers that integrate cleanly into ML pipelines.'''

from sklearn.impute import SimpleImputer, KNNImputer
from sklearn.pipeline import Pipeline

# Simple — mean/median/most_frequent
imputer = SimpleImputer(strategy="median")
df[["salary"]] = imputer.fit_transform(df[["salary"]])

# KNN — fills missing value based on K nearest neighbours
knn_imputer = KNNImputer(n_neighbors=5)
df_imputed = knn_imputer.fit_transform(df[["salary", "age", "experience"]])
```
